```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the image faces.jpg
img = cv2.imread("faces.jpg") # reads the image in BGR Format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
detector = cv2.CascadeClassifier("Datasets/FaceCascade/templatedata.xml")
```

```python
# after every round of checks, scale 1.3 times
faces = detector.detectMultiScale(img, 1.1)

for f in faces:
    x,y,w,h = f
    green = (0,255,0)
    cv2.rectangle(img, (x,y), (x+w,y+h), green, 5 )
    
plt.imshow(img)
plt.show()
```

```python
# create a Camera Object
camera = cv2.VideoCapture(0)

# read an Image
b,img = camera.read()
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Close the Camera
camera.release()

# display the image
plt.imshow(img)
```

```python
camera = cv2.VideoCapture(0)

while True:
    b,img = camera.read()
    if b==False:
        continue;
    
    # after every round of checks, scale 1.5 times
    faces = detector.detectMultiScale(img, 1.5)

    for f in faces:
        x,y,w,h = f
        green = (0,255,0)
        cv2.rectangle(img, (x,y), (x+w,y+h), green, 5 )

    cv2.imshow("Image", img)
      
    #     manage exit
    key = cv2.waitKey(1)&0xFF
    if key == ord('q'):
        break
        
    cv2.waitKey(1)
    
camera.release()
cv2.destroyAllWindows()
```
